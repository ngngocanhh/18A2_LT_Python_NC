```python
import numpy as np

# Giả lập dữ liệu
height = [70, 72, 68, 75, 71, 69, 74, 73, 70, 72] * 200
weight = [180, 190, 165, 210, 185, 170, 200, 195, 175, 188] * 200

#1
arr_height = np.array(height)
#2
arr_weight = np.array(weight)
#3
arr_height_m = arr_height * 0.0254
#4
arr_weight_kg = arr_weight * 0.453592
#5
arr_bmi = arr_weight_kg / (arr_height_m ** 2)
#6
print("Cân nặng tại index 50 (kg):", arr_weight_kg[50])
#7
print("Chiều cao từ index 100 đến 110 (m):", arr_height_m[100:111])
#8
print("BMI < 21:", arr_bmi[arr_bmi < 21])
#9
print("Chiều cao TB (m):", np.mean(arr_height_m))
print("Cân nặng TB (kg):", np.mean(arr_weight_kg))
#10
print("Chiều cao lớn nhất (m):", np.max(arr_height_m))
print("Cân nặng lớn nhất (kg):", np.max(arr_weight_kg))
#11
print("Chiều cao nhỏ nhất (m):", np.min(arr_height_m))
print("Cân nặng nhỏ nhất (kg):", np.min(arr_weight_kg))
```
