```python
import pandas as pd

# ================== TẠO DỮ LIỆU EURO 2012 ==================
data = {
    'Team': ['Germany', 'Italy', 'England', 'Spain', 'France', 'Greece', 'Russia'],
    'Goals': [10, 6, 5, 12, 3, 2, 5],
    'Yellow Cards': [4, 16, 5, 11, 6, 9, 7],
    'Red Cards': [0, 0, 0, 0, 0, 1, 1],
    'Shooting Accuracy': ['56%', '43%', '50%', '60%', '48%', '39%', '41%'],
    'Pass Accuracy': ['90%', '89%', '85%', '91%', '87%', '82%', '84%']
}

euro12 = pd.DataFrame(data)


print(" THÔNG TIN DATAFRAME EURO 2012 ")
print("Type:", type(euro12))
print("Shape:", euro12.shape)
print("Danh sách cột:", list(euro12.columns))

#1
print("\n1. Cột Goals:")
print(euro12['Goals'])

#2
print("\n2. Số đội tham gia Euro 2012:", euro12['Team'].nunique())

#3
print("\n3. Thông tin Euro 2012:")
print(euro12.info())

#4
discipline = euro12[['Team', 'Yellow Cards', 'Red Cards']]
print("\n4. DataFrame discipline:")
print(discipline)

#5
discipline_sorted = discipline.sort_values(
    by=['Red Cards', 'Yellow Cards'],
    ascending=False
)
print("\n5. Discipline sau khi sắp xếp giảm dần:")
print(discipline_sorted)

#6
#a
print("\n6a. Trung bình Yellow Cards:", euro12['Yellow Cards'].mean())
#b
print("\n6b. Các đội ghi hơn 6 bàn:")
print(euro12[euro12['Goals'] > 6])

#7
print("\n7. Đội có tên bắt đầu bằng 'G':")
print(euro12[euro12['Team'].str.startswith('G')])

#8
print("\n8. 7 cột đầu của euro12:")
print(euro12.iloc[:, :7])

#9
print("\n9. Tất cả các cột trừ 3 cột cuối:")
print(euro12.iloc[:, :-3])

#10
print("\n10. Các cột Team, Goals, Shooting Accuracy, Yellow Cards, Red Cards:")
print(euro12[['Team', 'Goals', 'Shooting Accuracy', 'Yellow Cards', 'Red Cards']])

#11
print("\n11. Shooting Accuracy của England, Italy, Russia:")
print(
    euro12.loc[
        euro12['Team'].isin(['England', 'Italy', 'Russia']),
        ['Team', 'Shooting Accuracy']
    ]
)
```
