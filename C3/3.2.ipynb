```python
#1
import numpy as np

arr = np.arange(10)

print("Array arr:", arr)
print("Kiểu dữ liệu:", arr.dtype)
print("Kích thước:", arr.shape)
```

```python
#2
arr_odd = arr[arr % 2 == 1]
arr_even = arr[arr % 2 == 0]

print("Array lẻ:", arr_odd)
print("Array chẵn:", arr_even)
```

```python
#3
arr_update_1 = arr.copy()
arr_update_1[arr_update_1 % 2 == 1] = 100

print("Array sau khi cập nhật:", arr_update_1)
```
