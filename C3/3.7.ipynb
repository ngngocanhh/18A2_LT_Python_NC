```python
import numpy as np

baseball = [
    [74, 180], [76, 215], [72, 210], [75, 205], [78, 230],
    [70, 185], [71, 190], [73, 200], [77, 220], [69, 175]    
] * 15          # nhân lên để đủ dữ liệu > 124 vận động viên
#1
np_baseball = np.array(baseball)

print("1. Kiểu dữ liệu:", np_baseball.dtype)
print("   Kích thước:", np_baseball.shape)

#2
print("\n2. Dòng thứ 50:", np_baseball[49])

#3
np_weight = np_baseball[:, 1]
print("\n3. Array cân nặng:", np_weight)

#4
print("\n4. Chiều cao VĐV thứ 124:", np_baseball[123, 0], "inch")

#5
print("\n5. Chiều cao TB:", np.mean(np_baseball[:, 0]), "inch")
print("   Cân nặng TB:", np.mean(np_baseball[:, 1]), "pound")

#6
corr = np.corrcoef(np_baseball[:, 0], np_baseball[:, 1])[0, 1]
print("\n6. Hệ số tương quan:", corr)
print(" => Có tương quan thuận giữa chiều cao và cân nặng")
```
