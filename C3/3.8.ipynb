```python
import numpy as np

heights = [
    190, 185, 178, 180, 182, 188, 175, 170, 176, 179,
    192, 187, 181, 174, 169, 183, 186, 177, 172, 184
]

positions = [
    'GK', 'D', 'M', 'A', 'M', 'GK', 'D', 'M', 'A', 'D',
    'GK', 'D', 'M', 'A', 'M', 'D', 'GK', 'A', 'M', 'D'
]

#1
np_positions = np.array(positions)
print("1a. Kiểu dữ liệu np_positions:", np_positions.dtype)

np_heights = np.array(heights)
print("1b. Kiểu dữ liệu np_heights:", np_heights.dtype)

#2
height_gk = np_heights[np_positions == 'GK']
print("\n2. Chiều cao trung bình GK:", np.mean(height_gk))

#3
height_not_gk = np_heights[np_positions != 'GK']
print("3. Chiều cao trung bình các vị trí khác GK:", np.mean(height_not_gk))

#4
players = np.zeros(len(np_heights), dtype=[
    ('position', 'U5'),
    ('height', 'float')
])

players['position'] = np_positions
players['height'] = np_heights

print("\n4. Mảng players:")
print(players)

#5
players_sorted = np.sort(players, order='height')

print("\n5. Cầu thủ thấp nhất:")
print(players_sorted[0])

print("Cầu thủ cao nhất:")
print(players_sorted[-1])
```
