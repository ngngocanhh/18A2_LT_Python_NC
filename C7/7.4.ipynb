```python
import tkinter as tk

def kiem_tra():
    # Nếu tất cả ô đều có dữ liệu thì hiện nút Gửi
    if o_ten.get() and o_id.get() and o_mat_khau.get():
        nut_gui.pack(pady=10)
    else:
        nut_gui.pack_forget()

cua_so = tk.Tk()
cua_so.title("Bài 7.4 - Form sinh viên")

# Nhãn và ô nhập Tên
tk.Label(cua_so, text="Tên sinh viên:").pack()
o_ten = tk.Entry(cua_so)
o_ten.pack()
o_ten.bind("<KeyRelease>", lambda e: kiem_tra())

# Nhãn và ô nhập ID
tk.Label(cua_so, text="ID sinh viên:").pack()
o_id = tk.Entry(cua_so)
o_id.pack()
o_id.bind("<KeyRelease>", lambda e: kiem_tra())

# Nhãn và ô nhập Mật khẩu
tk.Label(cua_so, text="Mật khẩu:").pack()
o_mat_khau = tk.Entry(cua_so, show="*")
o_mat_khau.pack()
o_mat_khau.bind("<KeyRelease>", lambda e: kiem_tra())

# Nút gửi (ẩn ban đầu)
nut_gui = tk.Button(cua_so, text="Gửi")

cua_so.mainloop()
```
