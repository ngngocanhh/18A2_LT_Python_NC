```python
import threading
import time

def web(name, delay):
    print(f"Starting {name}")
    for _ in range(3):
        time.sleep(delay)
        print(f"{name}: Web {time.ctime()}")
    print(f"Exiting {name}")

print("Starting Main Thread")

threads = [
    threading.Thread(target=web, args=("Google", 1)),
    threading.Thread(target=web, args=("Yahoo", 2)),
    threading.Thread(target=web, args=("Facebook", 3))
]

for t in threads:
    t.start()

for t in threads:
    t.join()

print("Main Thread")
```
