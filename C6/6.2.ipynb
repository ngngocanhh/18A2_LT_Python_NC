```python
import threading
import urllib.request

def download(url):
    response = urllib.request.urlopen(url)
    data = response.read()
    print(f"Đã tải {url} - {len(data)} bytes")

urls = [
    "https://www.example.com",
    "https://www.python.org",
    "https://www.wikipedia.org"
]

threads = []

for url in urls:
    t = threading.Thread(target=download, args=(url,))
    threads.append(t)
    t.start()

for t in threads:
    t.join()
```
