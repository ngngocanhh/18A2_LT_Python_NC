```python
import threading

limit = 20

def even_numbers():
    for i in range(0, limit + 1, 2):
        print("Chẵn:", i)

def odd_numbers():
    for i in range(1, limit + 1, 2):
        print("Lẻ:", i)

t1 = threading.Thread(target=even_numbers)
t2 = threading.Thread(target=odd_numbers)

t1.start()
t2.start()

t1.join()
t2.join()
```
