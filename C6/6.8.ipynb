```python
import threading
import random

lst = [random.randint(0, 10) for _ in range(40)]
num_threads = 4
chunk_size = len(lst) // num_threads
total_sum = 0
lock = threading.Lock()

def partial_sum(sublist, idx):
    global total_sum
    s = sum(sublist)
    with lock:
        total_sum += s
    print(f"Thread {idx}:", sublist, "=>", s)

threads = []
for i in range(num_threads):
    sub = lst[i*chunk_size:(i+1)*chunk_size]
    t = threading.Thread(target=partial_sum, args=(sub, i+1))
    threads.append(t)
    t.start()

for t in threads:
    t.join()

print("Tổng list:", total_sum)
```
