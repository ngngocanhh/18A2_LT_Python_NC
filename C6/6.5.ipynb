```python
import threading
import urllib.request

def send_request(url):
    response = urllib.request.urlopen(url)
    print(f"{url} -> Status OK")

urls = [
    "https://www.google.com",
    "https://www.yahoo.com",
    "https://www.facebook.com"
]

threads = []

for url in urls:
    t = threading.Thread(target=send_request, args=(url,))
    threads.append(t)
    t.start()

for t in threads:
    t.join()
```
