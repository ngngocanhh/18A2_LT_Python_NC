```python
class Stack:
    def __init__(self, capacity):
        self.capacity = capacity
        self.items = []

    def isEmpty(self):
        return len(self.items) == 0

    def isFull(self):
        return len(self.items) == self.capacity

    def push(self, item):
        if self.isFull():
            print("Ngăn xếp đã đầy!")
        else:
            self.items.append(item)

    def pop(self):
        if self.isEmpty():
            print("Ngăn xếp rỗng!")
            return None
        else:
            return self.items.pop()

    def count(self):
        return len(self.items)   
```
