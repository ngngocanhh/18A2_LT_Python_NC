```python
class PS:
    def __init__(self, ts, ms):
        print("Thông tin phân số:")
        self.ts = ts
        if ms == 0:
            print("Mẫu số không hợp lệ, gán mặc định = 1")
            self.ms = 1
        else:
            self.ms = ms

    def display(self):
        print("Tử số:", self.ts)
        print("Mẫu số:", self.ms)
        print("Phân số:", f"{self.ts}/{self.ms}")

    def hople(self):
        return self.ms != 0

ds = [PS(int(input("Tử số: ")), int(input("Mẫu số: "))) 
    for _ in range(int(input("Nhập số phân số: ")))]

print(" Danh sách phân số ")
for ps in ds:
    ps.display()
```
