```python
class TS:
    def __init__(self,HT,T,L,H):
        print('Thông tin của TS:')
        self.HT = HT
        self.T = T
        self.L = L
        self.H = H

    def display(self):
        print('Họ và tên:',self.HT)
        print('Điểm toán:',self.T)
        print('Điểm lý:',self.L)
        print('Điểm hóa:',self.H)
    
    def tongdiem(self):
        return self.T + self.L + self.H

ds = []
n = int(input("Nhập số lượng thí sinh: "))
for i in range(n):
    print(f"Thí sinh {i+1} ")
    ht = input("Họ và tên: ")
    t = float(input("Điểm Toán: "))
    l = float(input("Điểm Lý: "))
    h= float(input("Điểm Hóa: "))
    ts = TS(ht,t,l,h)
    ds.append(ts)

trung_tuyen = [ts for ts in ds if ts.tongdiem() >= 20]
trung_tuyen.sort(key=lambda x: x.tongdiem(), reverse=True)

print(" Danh sách thí sinh trúng tuyển:")
for ts in trung_tuyen:
    ts.display()
    print("Tổng điểm:", ts.tongdiem())

```
