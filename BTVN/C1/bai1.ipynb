```python
class HCN:
    def __init__(self,CD,CR,P,S):
        print('Các giá trị của HCN:')
        self.CD = CD
        self.CR = CR
        self.P = P
        self.S = S

    def display(self):
        print('Chiều dài:',self.CD)
        print('Chiều rộng:',self.CR)
        print('Chu vi:',self.P)
        print('Diện tích:',self.S)

Hinh_chu_nhat = HCN(27,14,82,378)
Hinh_chu_nhat.display()

```
